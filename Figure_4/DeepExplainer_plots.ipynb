```python
# With motif hits
import os
hits_directory = '/staging/leuven/stg_00002/lcb/cbravo/Liver/Multiome/DeepTopic/DARs_Zonation_lr_00001/individual_hits_with_scores_tfnnconv1d_meme/'
hits_files = os.listdir(hits_directory)
hits_files = [name for name in hits_files if 'bed' in name]

import pandas as pd

def format_info(hits_directory, file):
    info_df = pd.read_csv(hits_directory +file, sep='\t')
    tf = [file.split('_')[0]]*info_df.shape[0]
    start = list(info_df.hit_start-info_df.peak_start)
    end = list(info_df.hit_end-info_df.peak_start)
    score = list(info_df.hit_sum_scr)
    region_name = list(info_df.peak_chr+':'+info_df.peak_start.astype(str)+'-'+info_df.peak_end.astype(str))
    plot_df = pd.DataFrame([region_name, tf, start, end, score], index=['Region', 'TF', 'Start', 'End', 'Score']).T
    return plot_df

df_list = [format_info(hits_directory, hits_files[x]) for x in range(len(hits_files))]
hits_df = pd.concat(df_list)
```

```python
hits_df.replace('Fox', 'Foxa', inplace=True)
hits_df.replace('Onecut-1', 'Onecut', inplace=True)
hits_df.replace('Onecut-2', 'Onecut', inplace=True)
hits_df.replace('TAAT', 'Homeobox', inplace=True)
hits_df.replace('Tbx-1', 'Tbx', inplace=True)
hits_df.replace('Tbx-2', 'Tbx', inplace=True)
hits_df.replace('Tbx-3', 'Tbx', inplace=True)
hits_df.replace('Tbx-4', 'Tbx', inplace=True)
hits_df.replace('Tbx-dimer', 'Tbx', inplace=True)
hits_df.replace('Tcf7-1', 'Tcf7', inplace=True)
hits_df.replace('Tcf7-2', 'Tcf7', inplace=True)
hits_df.replace('Tcf7-3', 'Tcf7', inplace=True)
hits_df.replace('Tcf7-4', 'Tcf7', inplace=True)
hits_df.replace('Tcf7l1', 'Tcf7', inplace=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Patch
import random
import matplotlib.patches as mpatches
random.seed(555)
keys = ['Cebpa', 'Tbx', 'Hnf1a', 'Ppar', 'Hnf4a', 'Homeobox', 'Tcf7', 'Foxa', 'Stat', 'Nfi', 'Onecut', 'Creb']
values = ['deepskyblue', 'pink', 'gold', 'peru', 'forestgreen', 'grey', 'darkorange', 'red', 'darkgoldenrod', 'plum', 'indigo', 'turquoise']
color_dict=dict(zip(keys,values))

def plot_motif_hits(hits_df, region, color_dict, ntrack, track_no, fig):
    region_hits_df=hits_df[hits_df.Region == region].drop_duplicates()
    ax = fig.add_subplot(ntrack, 1, track_no)
    if region_hits_df.shape[0] > 0:
        for i in range(region_hits_df.shape[0]):
            ax.add_patch(Rectangle((list(region_hits_df.Start)[i],0), list(region_hits_df.End)[i]-list(region_hits_df.Start)[i], list(region_hits_df.Score)[i],color=color_dict[list(region_hits_df.TF)[i]], alpha=0.3, label=list(region_hits_df.TF)[i]))
    patch = [mpatches.Patch(color=color_dict[i], label=i) for i in color_dict.keys()]
    plt.legend(handles=patch)
    plt.xlim((0, 500))
    if region_hits_df.shape[0] > 0:
        plt.ylim((0, 0.01+np.max(list(region_hits_df.Score))))
    else: 
        plt.ylim((0, 0.01))
    return ax
```

```python
def get_plot_data(file, pos, flag):
    irf4_pos = []
    irf4_score = []
    irf4_pvalue = []
    with open(file,'r') as irf4file:
        for line in irf4file:
            if flag == 'sum':
                irf4_pos.append(int(line.strip().split('\t')[1])+pos)
                irf4_score.append(float(line.strip().split('\t')[3]))
                irf4_pvalue.append(float(line.strip().split('\t')[4]))
            if flag == 'no':
                irf4_pos.append(int(line.strip().split('\t')[1]))
                irf4_score.append(float(line.strip().split('\t')[7]))
                irf4_pvalue.append(float(line.strip().split('\t')[8]))
            if flag == 'ecr11':
                if int(splitted_line[1])-pos > -1:
                    irf4_pos.append(int(line.strip().split('\t')[1])-pos)
                    irf4_score.append(float(line.strip().split('\t')[3]))
                    irf4_pvalue.append(float(line.strip().split('\t')[4]))
    irf4 = {}
    with open(file,'r') as irf4file:
        for line in irf4file:
            splitted_line=line.strip().split('\t')
            if flag == 'sum':
                if int(splitted_line[1])+pos not in irf4:
                    irf4[int(splitted_line[1])+pos]= {'A':[], 'C':[], 'G':[], 'T':[]}
                #if splitted_line[3]=='-':
                #    irf4[int(splitted_line[1])][splitted_line[2]]=[float(splitted_line[7]),float(splitted_line[8])]
                if splitted_line[2]!='-':
                    irf4[int(splitted_line[1])+pos][splitted_line[2]]=[float(splitted_line[3]),float(splitted_line[4])]
            if flag == 'no':
                if int(splitted_line[1]) not in irf4:
                    irf4[int(splitted_line[1])]= {'A':[], 'C':[], 'G':[], 'T':[]}
                #if splitted_line[3]=='-':
                #    irf4[int(splitted_line[1])][splitted_line[2]]=[float(splitted_line[7]),float(splitted_line[8])]
                if splitted_line[3]!='-' and int(splitted_line[4])>=10:
                    irf4[int(splitted_line[1])][splitted_line[3]]=[float(splitted_line[7]),float(splitted_line[8])] 
            if flag == 'ecr11':
                if int(splitted_line[1])-pos not in irf4 and int(splitted_line[1])-pos > -1:
                    irf4[int(splitted_line[1])-pos]= {'A':[], 'C':[], 'G':[], 'T':[]}
                #if splitted_line[3]=='-':
                #    irf4[int(splitted_line[1])][splitted_line[2]]=[float(splitted_line[7]),float(splitted_line[8])]
                if splitted_line[2]!='-' and int(splitted_line[1])-pos > -1:
                    irf4[int(splitted_line[1])-pos][splitted_line[2]]=[float(splitted_line[3]),float(splitted_line[4])]
                
                    
    array_={'A':np.empty(500),'C':np.empty(500),'G':np.empty(500),'T':np.empty(500)}
    array_['A'][:]=np.nan
    array_['C'][:]=np.nan
    array_['T'][:]=np.nan
    array_['G'][:]=np.nan
    
    if flag == 'sum' or flag == 'ecr11':
        for i in range(500):
            if i in irf4:
                for key in irf4[i]:
                    if irf4[i][key]!=[]:
                        if irf4[i][key][1]<0.00001:
                            array_[key][i] = irf4[i][key][0]
    else:
        for i in range(500):
            if i+pos in irf4:
                for key in irf4[i+pos]:
                    if irf4[i+pos][key]!=[]:
                        if irf4[i+pos][key][1]<0.00001:
                            array_[key][i] = irf4[i+pos][key][0]  
                        
    array_['A'][array_['A']==0]=None
    array_['C'][array_['C']==0]=None
    array_['G'][array_['G']==0]=None
    array_['T'][array_['T']==0]=None
    return array_
```
